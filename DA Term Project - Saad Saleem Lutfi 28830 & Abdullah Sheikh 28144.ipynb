### Visualizing Income Groups

In this step, **monthly income** is prepared for analysis and converted into groups.

- Missing income values are replaced with the **median** to avoid dropping observations.
- Income is then divided into **three groups (Low, Medium, High)** using quantiles, ensuring each group contains a similar number of observations.
- This transformation makes income easier to analyze and compare across different demographic and socioeconomic variables.
- A bar chart is used to show the **proportion** of individuals in each income group.


```python
# Choose the column from your dataset
target_col = "Monthly_Income"

# Handle missing values (median is a safe choice for income)
df[target_col] = df[target_col].fillna(df[target_col].median())

# Create 3 income groups using quantiles (keeps groups balanced)
df["Income_Group"] = pd.qcut(df[target_col], q=3, labels=["Low", "Medium", "High"])

# Normalized value counts (proportions)
avg = df["Income_Group"].value_counts(normalize=True).sort_index()

# Bar plot
ax = avg.plot(kind="bar", ylim=(0, 1), title="Distribution of Income Groups", rot=0)

# Add labels on bars
ax.bar_label(ax.containers[0], fmt="%.2f")

plt.xlabel("Income Group")
plt.ylabel("Proportion")
plt.tight_layout()
plt.show()
```

#### Explanation:

The income distribution is relatively balanced across the three groups.

- Around **34%** of individuals fall into the **Low income** group.
- The **Medium income** group represents approximately **32%** of the sample.
- About **33%** of individuals are classified as **High income**.

This balanced distribution is expected because income groups were created using **quantiles**, which divide the data into groups of similar size.


### Visualizing a Binary Income Variable

After grouping income into three categories, income is further simplified into a **binary variable**.

- Individuals in the **Low income** group are classified as *Low*.
- Individuals in the Medium and High income groups are combined into *Not Low*.
- This binary classification allows for a clearer focus on **low-income outcomes** and simplifies later comparisons across demographic and socioeconomic variables.
- A bar chart is used to display the **proportion** of observations in each income class.


```python
# Binary target: Low income vs Not Low
df["Income_Binary"] = np.where(df["Income_Group"] == "Low", "Low", "Not Low")

avg_bin = df["Income_Binary"].value_counts(normalize=True)

ax = avg_bin.plot(kind="bar", ylim=(0, 1), title="Binary Income Classification", rot=0)
ax.bar_label(ax.containers[0], fmt="%.2f")

plt.xlabel("Class")
plt.ylabel("Proportion")
plt.tight_layout()
plt.show()
```

#### Explanation

The binary income classification shows a clear split between low-income and non-low-income individuals.

- Approximately **34%** of the sample is classified as **Low income**.
- The remaining **66%** falls into the **Not Low** category, which combines the medium and high income groups.

This result follows directly from the earlier quantile-based grouping and provides a simplified view of income status for further analysis.

Before cleaning, missing values were analyzed both in absolute numbers and percentages.  
Based on standard data cleaning practices discussed in lectures, the following rules were applied:

- Variables with **less than 5% missing values** were imputed using the **mode**, as the missingness was minimal.
- Variables with around **20% missing values** were retained if they were analytically important and imputed appropriately.
- Variables with **more than 50% missing values** were dropped due to the high risk of bias and information loss.

Based on this logic:
- **Education_Level** (≈21.8% missing) was retained and will be imputed due to its importance in income analysis.
- **Occupation** (≈3.7% missing) was imputed using the mode.

### Income by Education Level

This analysis compares monthly income across different education levels to understand how educational attainment is associated with income differences.

To improve readability, the income axis is restricted to focus on the majority of observations.

```python
# STEP 1: sanity check income column
df["Monthly_Income"] = pd.to_numeric(df["Monthly_Income"], errors="coerce")

print("Total rows:", len(df))
print("Non-missing Monthly_Income:", df["Monthly_Income"].notna().sum())
print("Min income:", df["Monthly_Income"].min())
print("Max income:", df["Monthly_Income"].max())

df["Monthly_Income"].describe()
```

```python
# STEP 7b: Inspect GeoJSON district names to fix remaining mismatches

# all district names available in map file
geo_names = sorted(gdf["district_key"].unique())

# helper: show geojson names that contain a substring
def find_geo(substring):
    substring = substring.lower()
    return [x for x in geo_names if substring in x]

targets = [
    "chagai", "duki", "kambar", "mirpur", "musakhel", "naushahro",
    "shaheed", "waziristan", "sujawal", "tando", "torghar", "umer",
    "east", "west"
]

for t in targets:
    matches = find_geo(t)
    print(f"\n--- matches for '{t}' ---")
    print(matches[:25])
```

## Constructing a New Variable
### Hourly Wage 
To make income comparisons more meaningful across individuals who work different hours, an hourly wage measure is created. This converts monthly income into an approximate hourly rate using reported weekly working hours.

Hourly wage is calculated as:

Hourly Wage = Monthly Income / (Hours Worked per Week × 4.33)

Hourly wage is only defined for individuals who reported working positive hours. For those who did not work during the week, hourly wage is left as missing.


```python
# Create hourly wage only for individuals who worked positive hours

df["Hourly_Wage"] = df["Monthly_Income"] / (df["Hours_Worked_Weekly"] * 4.33)

# Set hourly wage to missing where hours worked is zero or missing
df.loc[df["Hours_Worked_Weekly"] <= 0, "Hourly_Wage"] = None

# Re-check distribution
df["Hourly_Wage"].describe()
```

## Model Testing (Regression)

### Step 1: Confirm the target variable is clean and define X and y

Since the dataset was already cleaned earlier in this notebook, we do **not** drop missing values again.

Instead, we **verify** that `Monthly_Income` has no missing values and then define:
- `y` = Monthly Income (dependent variable)
- `X` = selected predictors (independent variables)

This keeps the workflow consistent and avoids repeating cleaning steps unnecessarily.


```python
# Step 1: Define y and X (based strictly on existing cleaned data)

target = "Monthly_Income"

feature_cols = [
    "Gender",
    "Age",
    "Marital_Status",
    "Education_Level",
    "Province",
    "District",
    "Region",
    "Occupation",
    "Hours_Worked_Weekly"
]

# Sanity check: ensure all features exist
missing_cols = [col for col in feature_cols if col not in df.columns]
print("Missing columns (should be empty):", missing_cols)

# Define X and y
X = df[feature_cols].copy()
y = df[target].copy()

print("Number of observations:", X.shape[0])
print("Number of features:", X.shape[1])

X.head()
```

### Step 2: Split the data into training and testing sets

To evaluate model performance fairly, the dataset is split into:
- **80% training data** (used to train the models)
- **20% testing data** (used only for evaluation)

The same train–test split will be used for **all regression models**, ensuring a valid comparison across models.


```python
from sklearn.model_selection import train_test_split

# Step 2: Train-test split (80/20)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training observations:", X_train.shape[0])
print("Testing observations:", X_test.shape[0])
```

### Step 3: Data preprocessing

The independent variables include both **numeric** and **categorical** features.

Before fitting regression models:
- Numeric variables are **standardized** so that they are on the same scale
- Categorical variables are **one-hot encoded** so that models can interpret them correctly

A `ColumnTransformer` is used to apply these transformations in a structured and reproducible way, ensuring that preprocessing is learned only from the training data.


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline

# Identify numeric and categorical columns
numeric_features = ["Age", "Hours_Worked_Weekly"]

categorical_features = [
    "Gender",
    "Marital_Status",
    "Education_Level",
    "Province",
    "District",
    "Region",
    "Occupation"
]

# Preprocessing pipelines
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

### Step 4: Train and evaluate a baseline Linear Regression model

We begin with **Linear Regression** as the baseline regression model.

This model helps us:
- Establish a reference performance level
- Compare whether more advanced models (Ridge, Lasso, Tree-based, ANN) improve predictive accuracy

The model is trained on the **training set** and evaluated on the **test set** using:
- **RMSE** (Root Mean Squared Error)
- **MAE** (Mean Absolute Error)
- **R²** (Explained variance)


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Step 4: Linear Regression model (baseline)

lin_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", LinearRegression())
])

# Train
lin_reg.fit(X_train, y_train)

# Predict on test set
y_pred_lr = lin_reg.predict(X_test)

# Evaluation metrics
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
mae_lr = mean_absolute_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Performance (Test Set)")
print("RMSE:", rmse_lr)
print("MAE:", mae_lr)
print("R-squared:", r2_lr)
```

### Step 5: Ridge Regression

Ridge Regression is an extension of Linear Regression that adds **L2 regularization**.

This penalty:
- Shrinks coefficient magnitudes
- Helps when predictors are highly correlated
- Often improves generalization performance

The same preprocessing pipeline and train–test split are used to ensure a fair comparison with the baseline Linear Regression model.


```python
from sklearn.linear_model import Ridge

# Step 5: Ridge Regression

ridge_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", Ridge(alpha=1.0))
])

# Train
ridge_reg.fit(X_train, y_train)

# Predict
y_pred_ridge = ridge_reg.predict(X_test)

# Evaluation metrics
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression Performance (Test Set)")
print("RMSE:", rmse_ridge)
print("MAE:", mae_ridge)
print("R-squared:", r2_ridge)
```

### Step 6: Lasso Regression

Lasso Regression applies **L1 regularization**, which can reduce some coefficients to zero.

This allows us to examine whether a simpler model, using fewer effective predictors, can perform as well as or better than Ridge and standard Linear Regression.

The same preprocessing pipeline and train–test split are maintained for a fair comparison.


```python
import numpy as np
import warnings
from sklearn.linear_model import Lasso
from sklearn.exceptions import ConvergenceWarning
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Suppress only the ConvergenceWarning for clean notebook output
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# Step 6: Lasso Regression (final, stable)
# Note: alpha increased to improve convergence on high-dimensional one-hot data

lasso_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", Lasso(
        alpha=0.05,        # stronger regularization -> easier convergence
        max_iter=200000,   # allow enough iterations
        tol=1e-4,          # reasonable tolerance for convergence
        random_state=42
    ))
])

# Train
lasso_reg.fit(X_train, y_train)

# Predict
y_pred_lasso = lasso_reg.predict(X_test)

# Evaluation metrics
rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Lasso Regression Performance (Test Set)")
print("RMSE:", rmse_lasso)
print("MAE:", mae_lasso)
print("R-squared:", r2_lasso)
```

## Step 6: Tree-Based Models (Regression)

We require tree-based methods in three forms:
1. **Tree Regressor (single tree)**
2. **Bagging (Random Forest)**
3. **Boosting (Gradient Boosting)**

### Step 6.1: Tree Regressor (Single Decision Tree)

We start with a **Decision Tree Regressor** as the simplest tree-based model.  
To avoid extreme overfitting, we use basic control parameters such as `max_depth` and `min_samples_leaf`.
We evaluate the model on the test set using:
- RMSE
- MAE
- R²


```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Step 6.1: Tree Regressor (Single Decision Tree)

tree_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", DecisionTreeRegressor(
        random_state=42,
        max_depth=15,        # limits complexity (reduces overfitting)
        min_samples_leaf=20  # prevents very small leaves
    ))
])

# Train
tree_reg.fit(X_train, y_train)

# Predict
y_pred_tree = tree_reg.predict(X_test)

# Metrics
rmse_tree = np.sqrt(mean_squared_error(y_test, y_pred_tree))
mae_tree = mean_absolute_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)

print("Tree Regressor (Decision Tree) Performance (Test Set)")
print("RMSE:", rmse_tree)
print("MAE:", mae_tree)
print("R-squared:", r2_tree)
```

### Step 6.2: Bagging – Random Forest Regressor

Bagging (Bootstrap Aggregating) improves upon a single decision tree by:
- training multiple trees on different bootstrap samples, and
- averaging their predictions to reduce variance and overfitting.

Random Forest is a standard bagging-based model and is well suited for complex,
non-linear income relationships.

We again evaluate the model on the test set using RMSE, MAE, and R².


```python
from sklearn.ensemble import RandomForestRegressor

# Step 6.2: Bagging (Random Forest Regressor)

rf_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", RandomForestRegressor(
        n_estimators=200,      # number of trees
        random_state=42,
        n_jobs=-1,             # use all cores
        max_depth=None,        # allow trees to grow fully
        min_samples_leaf=10    # slight regularization
    ))
])

# Train
rf_reg.fit(X_train, y_train)

# Predict
y_pred_rf = rf_reg.predict(X_test)

# Evaluation metrics
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regressor Performance (Test Set)")
print("RMSE:", rmse_rf)
print("MAE:", mae_rf)
print("R-squared:", r2_rf)
```

### Step 6.3: Boosting – Gradient Boosting Regressor

Boosting builds an ensemble of decision trees sequentially, where each new tree
focuses on correcting the prediction errors made by previous trees.

Gradient Boosting is particularly effective for structured tabular data and is
expected to perform well in predicting income, where complex non-linear patterns
and interactions may exist.

The model is evaluated on the test set using RMSE, MAE, and R².


```python
from sklearn.ensemble import GradientBoostingRegressor

# Step 6.3: Boosting (Gradient Boosting Regressor)

gbr_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", GradientBoostingRegressor(
        n_estimators=200,      # number of boosting stages
        learning_rate=0.1,     # standard learning rate
        max_depth=3,           # shallow trees to prevent overfitting
        random_state=42
    ))
])

# Train
gbr_reg.fit(X_train, y_train)

# Predict
y_pred_gbr = gbr_reg.predict(X_test)

# Evaluation metrics
rmse_gbr = np.sqrt(mean_squared_error(y_test, y_pred_gbr))
mae_gbr = mean_absolute_error(y_test, y_pred_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

print("Gradient Boosting Regressor Performance (Test Set)")
print("RMSE:", rmse_gbr)
print("MAE:", mae_gbr)
print("R-squared:", r2_gbr)
```

### Step 7: ANN (Artificial Neural Network)


Artificial Neural Networks can capture complex non-linear relationships by learning
multiple layers of transformations. This is useful in income prediction because the
relationship between variables (education, occupation, region, etc.) and income may
be non-linear and interactive.

We implement ANN using `MLPRegressor` (Multi-Layer Perceptron) from scikit-learn.
The same preprocessing pipeline (scaling + one-hot encoding) and the same 80/20
train–test split are used to ensure a fair comparison with previous models.

The ANN is evaluated on the test set using:
- RMSE
- MAE
- R²

```python
from sklearn.neural_network import MLPRegressor

# Step 7: ANN (MLPRegressor)

ann_reg = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", MLPRegressor(
        hidden_layer_sizes=(64, 32),  # two hidden layers
        activation="relu",
        solver="adam",
        alpha=0.0001,                 # L2 penalty (regularization)
        learning_rate_init=0.001,
        max_iter=300,
        random_state=42,
        early_stopping=True,          # prevents overfitting + speeds up training
        n_iter_no_change=10
    ))
])

# Train
ann_reg.fit(X_train, y_train)

# Predict
y_pred_ann = ann_reg.predict(X_test)

# Metrics
rmse_ann = np.sqrt(mean_squared_error(y_test, y_pred_ann))
mae_ann = mean_absolute_error(y_test, y_pred_ann)
r2_ann = r2_score(y_test, y_pred_ann)

print("ANN (MLPRegressor) Performance (Test Set)")
print("RMSE:", rmse_ann)
print("MAE:", mae_ann)
print("R-squared:", r2_ann)
```

After training all required machine learning models, we compare their performance
using the same evaluation metrics on the test set.

The comparison is based on:
